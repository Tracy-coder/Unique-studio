```python
#读取数据
PATH = './data/'
OUTPATH = './pro_data/'
df = pd.read_csv(PATH+"train.csv")
df1 = pd.read_csv(PATH+"test.csv")
# print(df.shape)
# print(df1.shape)
# print(df1)
print(df)
df.info()
#df1与df相比少了一列survived

df = df.drop("PassengerId",axis=1)
df = df.drop("Name",axis=1)
df = df.drop("Ticket",axis=1)

df1 = df1.drop("PassengerId",axis=1)
df1 = df1.drop("Name",axis=1)
df1 = df1.drop("Ticket",axis=1)

```

## 缺失值处理
+ 训练集中**age、cabin、embarked**这三列有缺失数据
+ 测试集中**age、cabin、fare**这三列有缺失数据

- cabin列缺的太多了 留下来意义不大 删掉！
- age、fare列用平均值填充
- embarked列用上一行的值填充



```python
print(df.isnull().any())
#age cabin和embarked这三列有缺失数据
print(df1.isnull().any())
#age cabin和fare这三列有缺失数据

#查看缺失了多少数据
print("df Age列")
print(df.Age.isnull().value_counts())
print("df Cabin列")
print(df.Cabin.isnull().value_counts())
print("df Embarked")
print(df.Embarked.isnull().value_counts())


print("df1 Age列")
print(df1.Age.isnull().value_counts())
print("df1 Cabin列")
print(df1.Cabin.isnull().value_counts())
print("df1 Fare列")
print(df1.Fare.isnull().value_counts())

# 丢弃cabin列
df = df.drop("Cabin",axis=1)
df1 = df1.drop("Cabin",axis=1)

# df的缺失值处理
df.loc[:,"Embarked"] = df["Embarked"].fillna(method='ffill')
df.loc[:,"Age"] = df.loc[:,"Age"].fillna(df["Age"].mean())
# df1的缺失值处理
df1.loc[:,"Age"] = df1.loc[:,"Age"].fillna(df1["Age"].mean())
df1.loc[:,"Fare"] = df1.loc[:,"Fare"].fillna(df1["Fare"].mean())
```

```python
# needcode_cat_columns = ["Pclass","Sex","SibSp","Parch","Embarked"]
# df_coded = pd.get_dummies(
# df,
# columns=needcode_cat_columns,
# prefix=needcode_cat_columns,
# dummy_na=True,
# drop_first=True
# )
# pd.get_dummies(df["Embarked"])

df.loc[:,"Sex"]=df["Sex"].str.replace("female","0")
df.loc[:,"Sex"]=df["Sex"].str.replace("male","1")

df.loc[:,"Embarked"]=df["Embarked"].str.replace("C","1")
df.loc[:,"Embarked"]=df["Embarked"].str.replace("S","2")
df.loc[:,"Embarked"]=df["Embarked"].str.replace("Q","3")

df1.loc[:,"Sex"]=df1["Sex"].str.replace("female","0")
df1.loc[:,"Sex"]=df1["Sex"].str.replace("male","1")

df1.loc[:,"Embarked"]=df1["Embarked"].str.replace("C","1")
df1.loc[:,"Embarked"]=df1["Embarked"].str.replace("S","2")
df1.loc[:,"Embarked"]=df1["Embarked"].str.replace("Q","3")

print(df)
print(df1)
# df = df_coded
```

### 归一化
age fare做归一化

```python
def Min_Max(series):
    max_value = series.max()
    min_value = series.min()
    return (series - min_value)/(max_value - min_value)

df["Age"] = Min_Max(df["Age"])
df["Fare"] = Min_Max(df["Fare"])
```
